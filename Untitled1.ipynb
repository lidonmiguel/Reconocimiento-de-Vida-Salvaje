## Introducción

En este notebook desarrollamos un sistema de reconocimiento de imágenes para clasificar fotografías en tres categorías: gatos (cat), perros (dog) y animales salvajes (wild). Utilizamos el dataset Animal Faces (AFHQ), que contiene miles de imágenes balanceadas para estas clases.

El modelo base es una red ResNet50 preentrenada con ImageNet, a la cual añadimos capas densas para adaptar la salida a nuestro problema de clasificación con 3 clases. Aplicamos técnicas de aumento de datos para mejorar la robustez del modelo y evitamos el sobreajuste mediante early stopping.

El objetivo final es conseguir una alta precisión en la clasificación y demostrar la capacidad del modelo para generalizar en un problema de reconocimiento visual complejo.

## Generador

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.resnet50 import preprocess_input

# Rutas a carpetas de train y val
train_dir = '/content/animal_faces/afhq/train'
val_dir = '/content/animal_faces/afhq/val'

# Generador con aumentos y preprocesamiento para entrenamiento
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)

# Generador solo con preprocesamiento para validación (sin aumentos)
val_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

# Generador de entrenamiento
# Generadores con tamaño reducido
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),  # ¡Cambio importante aquí!
    batch_size=32,
    class_mode='categorical'
)

# Generador de validación
validation_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(512, 512),
    batch_size=32,
    class_mode='categorical',
    shuffle=False  # Muy importante para evaluar bien
)
```

Entrenar

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Base del modelo
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False

# Capas superiores
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(3, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)
model.compile(optimizer=Adam(1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

## Entrenar

```python
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
checkpoint = ModelCheckpoint('mejor_modelo.keras', monitor='val_loss', save_best_only=True)

history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=20,
    callbacks=[early_stop, checkpoint]
)

# Guardar modelo después del ajuste
model.save('modelo.keras')
```

## Evaluar

```python
# Evaluar modelo
val_loss, val_acc = model.evaluate(validation_generator)
print(f"Precisión en validación: {val_acc*100:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Precisión (accuracy)
plt.plot(history.history['accuracy'], label='Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Validación')
plt.title('Precisión durante el entrenamiento')
plt.xlabel('Época')
plt.ylabel('Precisión')
plt.legend()
plt.show()

# Pérdida (loss)
plt.plot(history.history['loss'], label='Entrenamiento')
plt.plot(history.history['val_loss'], label='Validación')
plt.title('Pérdida durante el entrenamiento')
plt.xlabel('Época')
plt.ylabel('Pérdida')
plt.legend()
plt.show()
```

```python
import numpy as np

# Obtener predicciones en el conjunto de validación
Y_pred = model.predict(validation_generator)
y_pred = np.argmax(Y_pred, axis=1)
```

```python
# Etiquetas reales del generador
y_true = validation_generator.classes
class_names = list(validation_generator.class_indices.keys())
```

```python
from sklearn.metrics import classification_report

print("Reporte de clasificación:")
print(classification_report(y_true, y_pred, target_names=class_names))
```

## Ensayo

En esta sección realizamos un ensayo para probar el modelo entrenado con una imagen externa, no incluida en los datos de entrenamiento ni validación. Este proceso nos permite evaluar cómo se comporta el modelo ante datos reales y nuevos.

El procedimiento consiste en:

Cargar una imagen desde una ruta local.

Redimensionar la imagen al tamaño esperado por el modelo (224x224 píxeles).

Convertir la imagen a un array numérico y añadir una dimensión para simular un lote de tamaño 1.

Preprocesar la imagen con la misma función utilizada durante el entrenamiento para normalizarla correctamente.

Realizar la predicción con el modelo entrenado, obteniendo las probabilidades de pertenencia a cada clase.

Seleccionar la clase con mayor probabilidad y mostrar el nombre de la clase predicha (por ejemplo, gato, perro o salvaje).

Este ensayo es fundamental para verificar que el modelo funciona correctamente en escenarios prácticos y no sólo con los datos utilizados durante su desarrollo.



```python
import numpy as np
from tensorflow.keras.preprocessing import image

# Ruta a la imagen que quieres probar
img_path = '/content/perrito.jpg'  # Cambia por la ruta correcta

# Cargar la imagen y cambiar tamaño para que sea igual que la entrada del modelo
img = image.load_img(img_path, target_size=(224, 224))

# Convertir imagen a array y expandir dimensión para que sea batch=1
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)

# Preprocesar igual que durante entrenamiento
from tensorflow.keras.applications.resnet50 import preprocess_input
x = preprocess_input(x)

# Predecir clase
preds = model.predict(x)
predicted_class_index = np.argmax(preds, axis=1)[0]

# Mapa índice -> nombre clase
class_names = list(validation_generator.class_indices.keys())

print(f"La imagen se clasificó como: {class_names[predicted_class_index]}")
```
